```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000
sigma = 100
# Use large sigma so the vertical axis automatically generates more than two major ticks
# without me having to explicitly specify additional ticks.
# If we use sigma=1, the only major ticks will be at -10^0 and 10^0, which are also the min
# and max of the vertical axis. 

gaussian_samples = np.random.normal(0,sigma,N)
cumulative_sums = np.cumsum(gaussian_samples)
cumulative_averages = [cumulative_sums[i]/(i+1) for i in range(N)]
    
Ns = np.array(range(3, N+1)) # start at n=3 to prevent taking log of a negative number
upper_bound = sigma*np.sqrt(2*np.log(np.log(Ns))/Ns)
lower_bound = -upper_bound

plt.plot(Ns, cumulative_averages[2:], 'r', label='running average')
plt.plot(Ns, upper_bound, 'b', linewidth=3, label='LIL bounds')
plt.plot(Ns, lower_bound, 'b', linewidth=3)
plt.title('Running Average of Samples from N(0,1)')
plt.ylabel('Emprical Average')
plt.xlabel('Number of Samples in Emprical Average')
plt.xscale('log')
plt.yscale('symlog')
plt.ylim(-sigma,sigma)
plt.show()
```

```python
N = 10000
sigma = 100

num_trials = 3

for i in range(num_trials):
    gaussian_samples = np.random.normal(0,sigma,N)
    cumulative_sums = np.cumsum(gaussian_samples)
    cumulative_averages = [cumulative_sums[i]/(i+1) for i in range(N)]
    plt.plot(Ns, cumulative_averages[2:], label='running average ' + str(i))

Ns = np.array(range(3, N+1))
upper_bound = sigma*np.sqrt(2*np.log(np.log(Ns))/Ns)
lower_bound = -upper_bound


plt.plot(Ns, upper_bound, 'b', linewidth=3, label='LIL bounds')
plt.plot(Ns, lower_bound, 'b', linewidth=3)
plt.title('Running Average of Samples from N(0,1)')
plt.ylabel('Emprical Average')
plt.xlabel('Number of Samples in Emprical Average')
plt.xscale('log')
plt.yscale('symlog')
plt.ylim(-sigma,sigma)
plt.show()
```
